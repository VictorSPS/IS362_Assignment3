```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import pprint
import matplotlib.pyplot as plt
import matplotlib as mp
import random
```

```python
#I found on Google maps that from Baruch College in New York to Southernmost Key West point 1446 miles
total_miles = 1446
# from provided data we have
day1 = 55
day2 = 65
#with average daily distance we can find approximate qantity of days to travel
avg_daymiles=(day1+day2)/2
total_days=round(total_miles/avg_daymiles)

#Then, since we know only first two days of exact mileage, for the rest of the days I will generate random numbers in the range between 40 and 80 miles per day
distlist= [day1,day2]
daylist= ["day 1","day 2"]
i=0
day=2
while (day-2) != (total_days-2): 
    random_miles=random.randrange(40, 80)
    day= day+1
    distlist.append(random_miles)
    daylist.append("day "+str(day))
    summiles = sum(distlist)
    if summiles > 1371:
        continue
        distlist.append(1446 - sum(distlist))
print("Mileage for each day")
dataset= list(zip(daylist,distlist))
count=0
while count < day:    
    print(daylist[count],":",distlist[count],"miles")
    count = count+1
```

Then we will show cumulative miles

```python
cumulative=[]
cum=0
count=0
while cum <= total_miles:
    cum=cum+distlist[count]
    print (daylist[count],":",cum,"miles")
    cumulative.append(cum)
    count=count+1
    if count >= int(total_days):
        break
    
```
